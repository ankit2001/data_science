```python
%matplotlib inline
```

```python
x = np.arange(0, 20)
```

```python
y = x ** 2
```

```python
plt.plot(x, y)
```

```python
y = 3 * x + 9
```

```python
plt.plot(x, y)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'orange')
```

```python
plt.plot(x, y, 'g--')
```

```python
plt.xlim(0, 38)
plt.ylim(-38, 38)
y = np.sin(x)
plt.plot(x, y, 'o--')
```
